```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/credit_risk_dataset.csv")
```

# Find missing values

```python
df.isna().sum()
```

```python
df["person_emp_length"] = df["person_emp_length"].fillna(df["person_emp_length"].mean())
df["loan_int_rate"] = df["loan_int_rate"].fillna(df["loan_int_rate"].mean())
```

```python
df.isna().sum()
```

# Finding some descreptive stats about categorical variables

```python
cat_vars = []
num_vars = []
for i in df.columns:
    if df[i].dtype == object:
        cat_vars.append(i)
    else:
        num_vars.append(i)
```

```python
cat_vars
```

```python
def plot_bar(var1, var2, title):
    fig = plt.figure(figsize=(10,7))
    plt.bar(var1, var2)
    plt.title(title)
    plt.xticks(rotation = 45)
    plt.show()

for i in range(len(cat_vars)):
    var1 = list(df[cat_vars[i]].value_counts().keys())
    var2 = list(df[cat_vars[i]].value_counts().values)
    plot_bar(var1, var2, cat_vars[i])
```

```python
fig = plt.figure(figsize=(10,9))
sns.heatmap(df[num_vars].corr(), annot = True)
plt.show()
```

```python
def plot_hist(x, xlabel):
    fig = plt.figure(figsize=(10,9))
    plt.hist(x)
    plt.xlabel(xlabel)
    plt.title("Distribution of %s"%xlabel)
    plt.show()

for i in range(len(num_vars)):
    plot_hist(df[num_vars[i]], num_vars[i])
```

```python
person_home_ownership = pd.get_dummies(df["person_home_ownership"])
loan_intent = pd.get_dummies(df["loan_intent"])
loan_grade = pd.get_dummies(df["loan_grade"])
cb_person_default_on_file = pd.get_dummies(df["cb_person_default_on_file"])
```

```python
x_cols = list(df.columns)
x_cols.remove("loan_status")
X_ = df[x_cols]
X_.drop(columns = cat_vars, inplace = True)
X = pd.concat([X_, person_home_ownership, loan_intent, loan_grade, cb_person_default_on_file], axis = 1)
y = df["loan_status"]
```

```python
y.value_counts()
```

```python
X.columns
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
print("Before oversampling: ",Counter(y_train))
SMOTE= SMOTE()
X_train,y_train= SMOTE.fit_resample(X_train,y_train)
print("After oversampling: ",Counter(y_train))
```

```python
from sklearn.linear_model import LogisticRegression
logit= LogisticRegression()
logit.fit(X_train, y_train)
pred_logit= logit.predict(X_test)
```

```python
# a simple linear logistic model
import sklearn
model_adult = sklearn.linear_model.LogisticRegression(max_iter=10000)
model_adult.fit(X_train, y_train)
def model_adult_proba(x):
    return model_adult.predict_proba(x)[:,1]
def model_adult_log_odds(x):
    p = model_adult.predict_log_proba(x)
    return p[:,1] - p[:,0]
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_auc_score, plot_confusion_matrix, plot_precision_recall_curve

print("The accuracy of logit model is:", accuracy_score(y_test, pred_logit))
print(classification_report(y_test, pred_logit))
```
